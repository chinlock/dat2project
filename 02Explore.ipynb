### Step 2b: Companies and Jobs

```python
print 'Number of jobs', df['index'].size
```

```python
df1 = df['company'].where(df['company']!='').dropna()  # remove all entries where company is empty
print 'Number of unique companies', df1.unique().size
#df1.value_counts()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

# display plots in the notebook
%matplotlib inline

# plot the top 25 companies with most jobs
df1.value_counts().head(25).plot(kind='bar', alpha=0.5)
```

```python
print 'Number of unique job titles=', df.jobtitle.unique().size
#df['jobtitle'].value_counts()
```

```python
#plot the top 25 job titles
df['jobtitle'].value_counts().head(25).plot(kind='bar', alpha=0.5)
```
